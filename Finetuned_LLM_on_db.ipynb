```python
from huggingface_hub import HfApi, upload_folder
import os

# === Configuration ===
hf_username = "Yaso12345"  # Replace with your Hugging Face username
repo_name = "mistral-7b-lora-mysql"  # Your new or existing repo name
lora_folder_path = "/content/LLaMA-Factory/saves/Mistral-7B-Instruct-v0.3/lora/train_2025-06-11-07-41-45"  # Path to your LoRA adapter folder

# Construct full repo ID
repo_id = f"{hf_username}/{repo_name}"

# === Create the repo on HF if it doesn't exist ===
api = HfApi()

# Create repo (will do nothing if it already exists)
try:
    api.create_repo(repo_id=repo_id, repo_type="model", exist_ok=True)
    print(f"Repo created or already exists: {repo_id}")
except Exception as e:
    print(f"Failed to create repo: {e}")
    exit()

# === Upload the folder ===
print(f"Uploading LoRA adapter from '{lora_folder_path}' to Hugging Face...")

upload_folder(
    repo_id=repo_id,
    folder_path=lora_folder_path,
    commit_message="Upload LoRA adapter weights",
    path_in_repo=".",  # Upload everything as-is
    repo_type="model"
)

print(f"Upload complete! View it at: https://huggingface.co/{repo_id}")
```
